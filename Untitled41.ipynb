# 1. How many unique students are included in the dataset?

```python
unique_students = data['Email ID'].nunique()
print("Number of unique students:", unique_students)
```

# 2. What is the average GPA of the students?

```python
average_gpa = data['CGPA'].mean()
print("Average GPA of students:", round(average_gpa, 2))
```

# 3. What is the distribution of students across different graduation years?

```python
graduation_year_distribution = data['Year of Graduation'].value_counts()
print("\nDistribution of students across graduation years:\n", graduation_year_distribution)
```

# 4. What is the distribution of student's experience with Python programming?

```python
python_experience_distribution = data['Experience with python (Months)'].value_counts()
print("\nDistribution of Python programming experience:\n", python_experience_distribution)
```

# 5. What is the average family income of the students?

```python
# Mapping income ranges to numeric midpoints
income_mapping = {
    '0-2 Lakh': 1,
    '2-5 Lakh': 3.5,
    '5-7 Lakh': 6,
    '7 Lakh+': 8
}
data['Family Income (Lakhs)'] = data['Family Income'].map(income_mapping)
average_family_income = data['Family Income (Lakhs)'].mean()
print("\nAverage family income (Lakhs):", round(average_family_income, 2))
```

# 6. How does the GPA vary among different colleges? (Show top 5 results only)

```python
gpa_by_college = data.groupby('College Name')['CGPA'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 colleges by average GPA:\n", gpa_by_college)
```

# 7. Are there any outliers in the quantity (number of courses completed) attribute?

```python
quantiles = data['Quantity'].quantile([0.25, 0.5, 0.75])
iqr = quantiles[0.75] - quantiles[0.25]
lower_bound = quantiles[0.25] - 1.5 * iqr
upper_bound = quantiles[0.75] + 1.5 * iqr
outliers = data[(data['Quantity'] < lower_bound) | (data['Quantity'] > upper_bound)]
print("\nNumber of outliers in 'Quantity':", outliers.shape[0])
```

# 8. What is the average GPA for students from each city? (Show top 5 results)

```python
average_gpa_by_city = data.groupby('City')['CGPA'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 cities by average GPA:\n", average_gpa_by_city)
```

# 9. Can we identify any relationship between family income and GPA?

```python
family_income_gpa_corr = data[['Family Income (Lakhs)', 'CGPA']].corr().iloc[0, 1]
print("\nCorrelation between family income and GPA:", round(family_income_gpa_corr, 3))
```

# 14. How many students are graduating by the end of 2024? 

```python
# Assuming the 'Graduation Year' column exists
graduates_2024 = data[data['Year of Graduation'] == 2024].shape[0]
print(f'Total students graduating by the end of 2024: {graduates_2024}')
```

# 15. Which promotion channel brings in more student participations for the event?

```python
# Assuming 'data' is your DataFrame
promotion_channel_counts = data['How did you come to know about this event?'].value_counts()

# Select top 10
top_10_promotion_channels = promotion_channel_counts.head(10)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_promotion_channels.index, y=top_10_promotion_channels.values, palette="coolwarm")
plt.title('Top 10 Student Participation by Promotion Channel')
plt.xlabel('Promotion Channel')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.show()
```

# 16. Find the total number of students who attended the events related to Data Science?

```python
# Assuming Data Science events contain 'Data Science' in the event name
ds_event_attendance = data[data['Events'].str.contains('Data Science', na=False)].shape[0]
print(f'Total students who attended Data Science related events: {ds_event_attendance}')
```

# 17. Those who have high CGPA & more experience in language had high salary expectations? (Average)

```python
# High CGPA and Python experience
high_cgpa_python = data[(data['CGPA'] > data['CGPA'].mean()) & (data['Experience with python (Months)'] > data['Experience with python (Months)'].mean())]
avg_expected_salary = high_cgpa_python['Expected salary (Lac)'].mean()
print(f'Average expected salary for students with high CGPA and Python experience: {avg_expected_salary}')
```

# 18. How many students know about the event from their colleges? Which are the Top 5 colleges?

```python
# Students who learned about the event from their colleges
college_aware_students = data[data['How did you come to know about this event?'] == 'College Name']
top_5_colleges = college_aware_students['College Name'].value_counts().head(5)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_5_colleges.index, y=top_5_colleges.values, palette="magma")
plt.title('Top 5 Colleges with Most Students Aware of Event')
plt.xlabel('College Name')
plt.ylabel('Number of Students')
plt.show()

print(f'Total students who learned about the event from their college: {college_aware_students.shape[0]}')
```
